## Curvefit attempt 

```python
import scipy.optimize as sp
```

```python
def poly2(x,a,b,c):
    return a + b*x + c*x**2

def poly3(x,a,b,c,d):
    return a + b*x +c*x**2 + d*x**3

def poly3(x,a,b,c,d,e):
    return a + b*x +c*x**2 + d*x**3 + e*x**4

def poly4(x,a,b,c,d,e,f):
    return a + b*x +c*x**2 + d*x**3 + e*x**4 + f*x**5
```

```python
c32c = capacity32c*(32**2)
Tmin = 2.15
Tmax = 2.4
selection32c = np.logical_and(tempsls32c > Tmin, tempsls32c < Tmax)
peak_T32_values = np.array(tempsls32c)[selection32c]
peak_C32_values = c32c[selection32c]
```

```python
[a32c,b32c,c32c],error32 = sp.curve_fit(poly2,peak_T32_values,peak_C32_values)
```

```python
fig = plt.figure(figsize=(5,5))
cap1 = fig.add_subplot(1,1,1)
cap1.plot(tempsls32c,capacity32c*(32**2), c = 'blue',linestyle='',marker='.')
cap1.set_ylabel(r"Specfic Heat capacity / $k_{b}\ JK^{-1}$ ") #correct units?
cap1.set_xlabel("Temperature / K")
x2_32c = np.linspace(Tmin,Tmax)
y2_32c = []
for i in x2_32c:
    y2_32c.append(poly2(i,a32c,b32c,c32c))
    
cap1.plot(x2_32c,y2_32c)
plt.savefig('hc3232C_fit.svg',bbox_inches='tight')
```
